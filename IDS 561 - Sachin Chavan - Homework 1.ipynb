```python
import pandas as pd
import numpy as np
#reading the file temperatures.txt
with open ('C:/Users/Sachin Chavan/Downloads/temperatures.txt') as f:
    rows = f.readlines() #adding the content of the file to a variable
read = [] #empty list to add the output of our functions
for i in range(len(rows)): 
    read.append(rows[i].strip('\n')) #passing the contents of the variable to the empty list
```

```python
import string
def cleaning_data(read): #function for data cleaning part, passing 'read' as input
    read_temp = [] #creating an empty list to store result
    for str in read:
        line = str.strip() #removing spaces at the beginning & ending of the string
        line = str.translate(str.maketrans('','',string.punctuation)) #removing punctuation marks from the string
        line = [[int(s) for s in line.split(" ")]] #converting string to integer
        read_temp.append(line)      
    return read_temp
#end of cleaning function
df = cleaning_data(read) 
df
```

```python
#function for splitting dataset into two parts, first part includes first 500 lines & second includes rest 500 lines
#passing input of cleaning_data function to the split function
def splitting_data(df):
    P1 = df[0:500] #first part includes first 500 lines
    P2 = df[500:1000] #second part includes the rest 500 lines
    return P1, P2 #returning the two parts
P1,P2 = splitting_data(df) #passing the two split parts back to df, calling function
```

```python
# function for mapping data & creating set of key-value pairs for P1, P2, passing 'read' as input
def map(read):
    date = []  #list to append date values
    tempe = [] #list to append tempertaure values
    result=[]  #list to store the result
    final_temp = []
    for a in read: #we need to use three loops as there are three lists we need to get traverse to get to the data, here we get into the first list 
        for b in a: #getting inside the second list
            for c in range(len(b)): #getting inside the final list, while iterating through the indexes
                if c%2==0:  #as dates are placed on even indices like 0,2,4....
                    date.append(b[c]) #pulling dates from the list and appending to the date list formed above
                else:
                    tempe.append(b[c]) #appending the rest of the data on odd indices to the temperature list
    
    for a in range(len(date)):
        result.append({'Date':date[a],'Temperature':tempe[a]}) #appending the date and temperatures to the result list
    return result   
#end of map function
#taking P1, P2 as input from the splitting_data function
map_p1 = map(P1) #passing P1 through the map function
map_p2 = map(P2) #passing P2 through the map function

#adding both to create one consolidated list and storing in final_temperature variable
final_temperature = map_p1 + map_p2 

#printing the output
print(final_temperature)
```

```python
# function for sorting the data from map_p1 and map_p2
def sorting(sorting_data):
    sorted_list = sorted(sorting_data, key = lambda d : d['Date'], reverse = False)  #sorting the data on the basis of 'Date' parameter, also the reverse flag is set to 'False' to ensure ascending sort
    return sorted_list
#end of sorting function

#passing the output of the map function to the sorting function and storing it in sort_temp variable
z = sorting(final_temperature)

#printing the sorted list
print(z)
```

```python
#In order to partion data into pre and post 2015 phase we'll need to extract the year from the 'Date' data

df = pd.DataFrame(z) #creating a dataframe and passing the output of the sorting function
df['Date'] = df['Date'].astype(str) #converting to string to extract sub-string of years 
df['Year'] =[z[0:4] for z in df['Date']] #extracting first 4 numbers from the date column which represent years and saving it under 'Year'
df['Date'] = df['Date'].astype(int)
df['Year'] = df['Year'].astype(int)
df.head()
```

```python
#Partioning function to split years into pre 2015 to be sent to reducer 1 and post 2015 to be sent to reducer 2
def partition(p):   
    dp1 = p[p['Year'] <= 2015] #splitting into pre-2015
    dp2 = p[p['Year'] > 2015] #splitting into post-2015
    return dp1,dp2
#end of partition function
part_1,part_2 = partition(df) #passing the partions back to the dataframe
w = partition(df)
```

```python
w #checking if the partition has been done correctly
```

```python
#Need to define a reducer function that will find the maximum temperature from the two partions individually. We will pass the two partitions separately
def reducer(output):
   
    reduce_data = output.groupby(['Year'], as_index = False)['Temperature'].agg(list) #grouping year and temperature values together and leaving out the Date values altogether
    maximum = [] #an empty list to store the maximum temperature values     
    for i in range(len(reduce_data)): #loop to find the maximum value in a partition
        maximum.append(max(reduce_data['Temperature'][i]))  
    reduce_data['Maximum Temperature'] = maximum #passing the max value corresponding years to the empty 'maximum' list  
    return reduce_data

#passing both the partitions through the reducers
max_temp1 = reducer(part_1)
max_temp2 = reducer(part_2)
```

```python
#Deleting the temperature column to remove the year column & only keep the maximum temperature
del max_temp1['Temperature']
del max_temp2['Temperature']
#adding both the partitions to ensure consolidated data
final_maxtemp = pd.concat([max_temp1, max_temp2])
```

```python
ft = final_maxtemp.reset_index(drop=True) #correcting the index number order
ft #checking the final output
```

```python
def main(read):
    
    #data cleaning function
    df = cleaning_data(read)
    
    #data split function
    P1,P2 = splitting_data(df)
    
    #data map function
    map_p1 = map(P1)
    map_p2 = map(P2)
    final_temperature = map_p1 + map_p2
    
    #data sort function, calling the variable where the sorting function is stored
    z 
    
    #data partition function, calling the variable where partion(df) is stored
    w
    
    #data reducer function
    max_temp1 = reducer(part_1)
    max_temp2 = reducer(part_2)
    
    #deleting the unnecessary temperature column
    del max_temp1['Temperature']
    del max_temp2['Temperature']
    
    final_maxtemp = pd.concat([max_temp1, max_temp2])
    ft = final_maxtemp.reset_index(drop=True)
    
    return ft
```

```python
print(main(read)) #printing the final output
```
